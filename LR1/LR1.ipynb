```python
sns.set_style("white")

# Basic 2D density plot
sns.kdeplot(x=dat.Age, y=dat.Length)
plt.show()

# Custom the color, add shade and bandwidth
sns.kdeplot(x=dat.Age, y=dat.Length, cmap="Reds", shade=True, bw_adjust=.5)
plt.show()

# Add thresh parameter
sns.kdeplot(x=dat.Age, y=dat.Length, cmap="Blues", shade=True, thresh=0)
plt.show()
```
